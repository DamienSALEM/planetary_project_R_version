```r
cols_to_keep <- c("P_HABITABLE", "P_HABZONE_OPT", "P_ESI", "P_HABZONE_CON",
                  "S_LOG_LUM_ERROR_MAX", "S_LOG_G",
                  "P_TEMP_SURF", "P_TEMP_EQUIL",
                  "S_LOG_LUM", "S_MASS", "S_DISTANCE")


data_ia_second <- data_ia[, cols_to_keep]
head(data_ia_second)
```

```r
#Remove line with more than 4 NULL values
str(data_ia_second)
print(paste("Nombre de lignes :", nrow(data_ia_second)))
na_count_per_row <- apply(data_ia_second, 1, function(x) sum(is.na(x)))

data_ia_second <- data_ia_second[na_count_per_row <= 4, ]

str(data_ia_second)

print(paste("Nombre de lignes restantes:", nrow(data_ia_second)))
```

```r
columns_to_mean <- c("P_ESI", "S_LOG_LUM_ERROR_MAX", "S_LOG_LUM", "P_TEMP_EQUIL", "S_MASS")
for (col in columns_to_mean) {
  data_ia_second[[col]][is.na(data_ia_second[[col]])] <- mean(data_ia_second[[col]], na.rm = TRUE)
}

cor(data_ia_second$P_TEMP_SURF, data_ia_second$S_DISTANCE, use = "complete.obs")

if (!all(is.na(data_ia_second$P_TEMP_SURF))) {
  model <- lm(P_TEMP_SURF ~ S_DISTANCE, data = data_ia_second, na.action = na.exclude)
  predicted_temps <- predict(model, newdata = data_ia_second[is.na(data_ia_second$P_TEMP_SURF), ])
  data_ia_second$P_TEMP_SURF[is.na(data_ia_second$P_TEMP_SURF)] <- predicted_temps
} else {
  data_ia_second$P_TEMP_SURF[is.na(data_ia_second$P_TEMP_SURF)] <- mean(data_ia_second$P_TEMP_SURF, na.rm = TRUE)
}

summary(data_ia_second)
```

```r
data_ia_second$P_HABITABLE <- ifelse(data_ia_second$P_HABITABLE == 2, 1, 0)

table(data_ia_second$P_HABITABLE)
```

```r
# Installer et charger les packages nécessaires
if (!requireNamespace("caret", quietly = TRUE)) {
    install.packages("caret")
}
library(caret)

if (!requireNamespace("randomForest", quietly = TRUE)) {
    install.packages("randomForest")
}
library(randomForest)

# Fixer la graine aléatoire pour la reproductibilité
set.seed(123)

# Préparer les données
# Assurez-vous que toutes les autres variables à part P_HABITABLE sont utilisées comme features
features <- setdiff(names(data_ia_second), "P_HABITABLE")

# Créer des indices pour diviser les données en train et test
trainIndex <- createDataPartition(data_ia_second$P_HABITABLE, p = 0.8, list = FALSE)

# Séparer les données en ensemble d'entraînement et de test
trainData <- data_ia_second[trainIndex, features]
testData <- data_ia_second[-trainIndex, features]

# Séparer les étiquettes cibles pour l'entraînement et le test, et les convertir en facteur pour la classification
trainLabels <- as.factor(data_ia_second[trainIndex, "P_HABITABLE"])
testLabels <- as.factor(data_ia_second[-trainIndex, "P_HABITABLE"])

# Gestion des données manquantes - Imputation par la médiane pour les variables continues
for(feature in features) {
  if(is.numeric(trainData[[feature]])) {
    median_value <- median(trainData[[feature]], na.rm = TRUE)
    trainData[[feature]][is.na(trainData[[feature]])] <- median_value
    testData[[feature]][is.na(testData[[feature]])] <- median_value
  }
}

tuneGrid <- expand.grid(mtry = 2:4)

fitControl <- trainControl(method = "cv", number = 10)

model <- train(x = trainData, y = trainLabels,
               method = "rf",
               ntree = 500,
               tuneGrid = tuneGrid,
               trControl = fitControl)

predictions <- predict(model, newdata = testData)

# Évaluer la performance du modèle
confMatrix <- confusionMatrix(predictions, testLabels)
print(confMatrix)

# Afficher les résultats du meilleur modèle
print(model)
```
